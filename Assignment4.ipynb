```python
import numpy as np
import IPython.display as ipd
```

```python
def getFrequency(note):
    A4=440
    notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']

    octave = int(note[2]) if len(note) == 3 else int(note[1])
        
    keyNumber = notes.index(note[0:-1]);
    
    if (keyNumber < 3) :
        keyNumber = keyNumber + 12 + ((octave - 1) * 12) + 1; 
    else:
        keyNumber = keyNumber + ((octave - 1) * 12) + 1; 

    return A4 * 2** ((keyNumber- 49) / 12)
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

C5  = getFrequency('C5')
D5  = getFrequency('D5')
E5  = getFrequency('E5')
F5  = getFrequency('F5')
G5  = getFrequency('G5')
A5 = getFrequency('A5')
B5 = getFrequency('B5')
C6 = getFrequency('C6')
D6 = getFrequency('D6')
E6 = getFrequency('E6')
F6 = getFrequency('F6')
G6 = getFrequency('G6')
```

```python
# "날아라 병아리 (by. N.EX.T)"

notes = [E5, G5, F5, E5, D5, C5, E5, 0, \
         E5, G5, F5, E5, D5, C5, E5, D5, C5, 0, \
         E5, G5, F5, 0, F5, A5, G5, D6, C6, D6, Db6, E6, \
         F6, G6, F6, E6, D6, C6, E6, D6, C6]

durs = [0.5, 0.25, 0.75, 0.25, 0.75, 0.25, 1.00, 0.25, \
        0.5, 0.25, 0.75, 0.25, 0.75, 0.5, 0.25, 0.25, 0.5, 0.25, \
        0.5, 0.25, 0.75, 0.25, 0.5, 0.25, 0.75, 0.5, 0.5, 0.75, 0.5, 0.75, \
        0.5, 0.25, 0.75, 0.25, 0.75, 0.5, 0.25, 0.25, 0.5]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
