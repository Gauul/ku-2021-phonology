```python
import matplotlib.pyplot as plt
import pandas as pd
```

1. pitch plot

```python
df = pd.read_csv('/content/yourgoodpants.csv', sep=',')
print(df)
```

```python
df.loc[df.Time_s == "--undefined--", "Time_s"] = None
df.loc[df.F0_Hz == "--undefined--", "F0_Hz"] = None
print(df)
```

```python
print(df.dtypes)
```

```python
time = df.Time_s
pitch = df.F0_Hz.astype(float)

plt.plot(time, pitch, 'bo')
plt.xlabel('time(sec)')
plt.ylabel('pitch(Hz)')
plt.show()
```

2. intensity plot

```python
df = pd.read_csv('/content/yourgoodpants_intensity.csv', sep=',')
print(df)
```

```python
time = df.Time_s
intensity = df.intensity_dB

plt.plot(time, intensity, 'yo')
plt.xlabel('time(sec)')
plt.ylabel('intensity(dB)')
plt.show()
```

3. formants plot

```python
df = pd.read_csv('/content/yourgoodpants_formants.csv', sep=',')
print(df)
```

```python
time = df.Time_s
f1 = df.F1_Hz
f2 = df.F2_Hz
f3 = df.F3_Hz
f4 = df.F4_Hz

plt.plot(time, f1, 'ro', label='F1')
plt.plot(time, f2, 'bo', label='F2')
plt.plot(time, f3, 'go', label='F3')
plt.plot(time, f4, 'yo', label='F4')
plt.xlabel('time(sec)')
plt.ylabel('frequency(Hz)')
plt.legend(loc='best', ncol=2, fontsize=10)
plt.show()
```
